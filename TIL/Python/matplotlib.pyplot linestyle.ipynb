# matplotlib.pyplot linestyle, color

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(10)


fig, ax = plt.subplots()

#solid
ax.plot(x,x,linestyle='-')

#dashed
ax.plot(x,x+2,linestyle='--')

#dashdot
ax.plot(x,x+4, linestyle='-.')

#dotted
ax.plot(x,x+6, linestyle=':')
```

## practice

```python
fig, ax =plt.subplots()

x= np.arange(15)
y= x**2

ax.plot(x,y,
       linestyle=':',
       marker='*',
       color='#524Fa1')
```

## color

```python
x = np.arange(10)
fig, ax =plt.subplots()

ax.plot(x,x, color ='r')

ax.plot(x,x+2, color='green')

ax.plot(x,x+4, color='0.8')

ax.plot(x, x+6, color ='#524FA1')
```

```python
x = np.arange(10)
fig, ax =plt.subplots()

ax.plot(x,x, color ='blue')
```

## marker

```python
x = np.arange(10)

fig, ax =plt.subplots()
ax.plot(x,x, marker='.')
ax.plot(x,x+2, marker='o')
ax.plot(x,x+4, marker='v')
ax.plot(x,x+6, marker='s')
ax.plot(x,x+8, marker='*')
```
