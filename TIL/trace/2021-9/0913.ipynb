```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('19년.csv')
```

```python
df['주소'].isnull().sum()
```

```python
df=df.dropna(subset=['주소'])
```

```python
df['주소'].isnull().sum()
```

```python
df=df[df['주소'].str.contains('종로구')]
```

```python
df.shape
```

```python
df.to_csv('종로구_교통량장소.csv',index=False,encoding='cp949')
```

```python
df_location=df[['위도','경도']]
df_location
```

```python
import folium
df['위도'].mean() #위도의 평균값
df['경도'].mean() #경도
```

```python
map=folium.Map(location=[df['위도'].mean(),df['경도'].mean()],zoom_start=12)


for n in df.index:
    name=df.loc[n,'지점명칭']
    
    location=[df.loc[n,'위도'], df.loc[n,'경도']]
#     adress=df.loc[n,'구']
    popup=f'{name}'

    folium.Marker(
        location = location,
        popup = popup,
    ).add_to(map)
map
```
