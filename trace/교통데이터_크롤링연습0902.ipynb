```python
url = 'http://data.ex.co.kr/openapi/trafficapi/trafficAll'
key = '7467044539'

url+= '?key='+key
url+= '&type=xml'

response = requests.get(url)
print(response.status_code)
print(response.text)
```

```python
if response.status_code==200:
    xd = xmltodict.parse(response.text)
    xdj = json.dumps(xd, ensure_ascii = False)
    xdjd = json.loads(xdj)
```
